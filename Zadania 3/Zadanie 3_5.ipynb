```python
stop=input('Skolko chisel hohite vvesty? ')
l1=list()
for i in range(int(stop)):
    i=int(input('Vvedite chislo: '))
    l1.append(i)
def defrel(n):
    tuple1=list()
    k = 0
    for i in range(2, n+1):
   
        for j in range(2, i):
       
            if i % j == 0:
             k = k + 1
  
        if k == 0:
            tuple1.append(i)
        else:
            k = 0
    return tuple1
def dict1(l):
    d=dict.fromkeys(l)
    for i in d.keys():
        d[i] = defrel(i)
    return d


print (dict1(l1))
```
